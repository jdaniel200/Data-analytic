```python
import pandas as pd
import re
import openpyxl
from google.colab import files
uploaded = files.upload()

# Función para convertir DMS a decimal
def dms_to_decimal(dms):
    # Expresión regular para capturar grados, minutos, segundos (con decimales) y la dirección
    dms_pattern = re.compile(r"(\d+)°(\d+)'([\d\.]+)\"?\s*([NSEW])")
    match = dms_pattern.match(dms)

    if match:
        degrees = int(match.group(1))
        minutes = int(match.group(2))
        seconds = float(match.group(3))  # Aquí permitimos decimales en los segundos
        direction = match.group(4)

        # Convertir a decimal
        decimal = degrees + minutes / 60 + seconds / 3600
        if direction in ['S', 'W']:
            decimal = -decimal
        return decimal
    else:
         print(f"Formato DMS inválido: {dms}")
         return None

# Cargar el archivo de Excel
df = pd.read_excel('coordenadas_1.xlsx')

# Asumiendo que las coordenadas están en dos columnas: 'Latitud' y 'Longitud'
df['Latitud Decimal'] = df['Latitud'].apply(dms_to_decimal)
df['Longitud Decimal'] = df['Longitud'].apply(dms_to_decimal)

# Guardar como archivo csv
df.to_csv('coordenadas_decimal.csv', index=False)


print("Conversión completada. Archivo guardado como 'coordenadas_decimal.csv'")
```

```python
import pandas as pd
import re
import openpyxl
from google.colab import files
uploaded = files.upload()


# Función corregida para convertir de GMS (grados, minutos, segundos) a decimal
def gms_to_dec(coord):
    # Expresión regular para separar grados, minutos, segundos y dirección
    elem = re.split(r"[°'\"]+", coord)  # Separar usando símbolos de grados, minutos y segundos

    # Convertir grados, minutos y segundos a valores numéricos

    dec = int(elem[0]) + int(elem[1]) / 60 + float(elem[2]) / 3600

    # Suma para obtener la coordenada en formato decimal


    # Identificar la dirección cardinal (N/S/E/W)
    if 'S' in coord or 'W' in coord:
        dec = -dec

    return dec

# Cargar el archivo de Excel
df = pd.read_excel('coordenadas_1.xlsx')

# Asumiendo que las coordenadas están en dos columnas: 'Latitud' y 'Longitud'
df['Latitud Decimal'] = df['Latitud'].apply(gms_to_dec)
df['Longitud Decimal'] = df['Longitud'].apply(gms_to_dec)

# Guardar como archivo csv
df.to_csv('coordenadas_decimal.csv', index=False)


print("Conversión completada. Archivo guardado como 'coordenadas_decimal.csv'")
```
